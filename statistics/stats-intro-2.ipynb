```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
dataset = np.array([7,3,5,6,4,8,9,2,1,-275,275])
```

```python
def detect_outliers(data):
    _outliers = np.empty(shape=(0),dtype=np.int64)
    _thrreshold = 1.5
    mean = np.mean(data)
    std = np.std(data)
    for i in data:
        z_score = (i-mean)/std
        if(np.abs(z_score) >= _thrreshold):
            _outliers = np.append(_outliers,i)
    return _outliers
```

```python
plt.hist(dataset)
```

```python
outliers = detect_outliers(dataset)
print("outliers = ", outliers)
```

```python
def get_q1_q3(data):
    return np.percentile(data, [25,75])
```

```python
q = get_q1_q3(dataset)
iqr = q[1] - q[0]
print("iqr = ", iqr)
print("lower fence = ", q[0] - 1.5*(iqr))
print("upper fence = ", q[1] + 1.5*(iqr))
```
